The entity-like to convert must be passed as keyword argument.

```python
from_compatible("SpontaneousMagnetization", "kA/m", Ms=me.Ms())
```

```python
from_compatible("ThermodynamicTemperature", "deg_C", T=me.T(), Ms=me.Ms())  # mistake of the developer
```

All additional options must be given as keyword arguments to avoid ambiguity.

```python
from_compatible("ThermodynamicTemperature", "deg_C", enforce_unit=True, temperature=me.T())
```

```python
from_compatible("ThermodynamicTemperature", "deg_C", temperature=me.T().q)
```

```python
from_compatible("ThermodynamicTemperature", "deg_C", temperature=600)
```

```python
from_compatible("ThermodynamicTemperature", "deg_C", temperature=me.Tc(80))  # mistake of the user
```

```python
from_compatible("ThermodynamicTemperature", "deg_C", compatible_entities=("CurieTemperature",), temperature=me.Tc(80))
```

```python
from_compatible("ThermodynamicTemperature", "deg_C", compatible_entities=("CurieTemperature",), temperature=me.Tc(80), enforce_unit=True)
```

Usage error: declaring incompatible entities as compatible:

```python
from_compatible("ThermodynamicTemperature", "deg_C", compatible_entities=("SpontaneousMagnetization",), temperature=me.Ms())  # mistake of the developer
```

```python
class Results(me.EntityCollection):
    def __init__(self, Ms, Tc):
        ensure_entity("Ms", Ms, "SpontaneousMagnetization")
        ensure_entity("Tc", Tc, "CurieTemperature")
        super().__init__(Ms=Ms, Tc=Tc)
```

```python
Results()
```

```python
Results(1, 2)
```

```python
Results(me.Ms(), me.Ms())
```

```python
res.A = me.A()
res
```

```python
del res.Ms
res
```
