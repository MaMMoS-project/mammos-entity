```python
import mammos_entity as me
```

```python
from mammos_entity._entity import ensure_entity
from mammos_entity._entity_collection import frozen_collection
```

```python
class Results(me.EntityCollection):
    def __init__(self, Ms, Tc):
        ensure_entity("SpontaneousMagnetization", Ms=Ms)
        ensure_entity("CurieTemperature", Tc=Tc)
        super().__init__(Ms=Ms, Tc=Tc)
```

```python
Results()
```

```python
Results(1, 2)
```

```python
Results(me.Ms(), me.Ms())
```

```python
res = Results(me.Ms(), me.Entity("CurieTemperature"))
res
```

```python
res.A = me.A()
res
```

```python
del res.Ms
res
```

```python
@frozen_collection
class ResultsFrozen(me.EntityCollection):
    def __init__(self, Ms, Tc):
        ensure_entity("SpontaneousMagnetization", Ms=Ms)
        ensure_entity("CurieTemperature", Tc=Tc)
        super().__init__(Ms=Ms, Tc=Tc)
```

```python
res_frozen = ResultsFrozen(me.Ms(), me.Entity("CurieTemperature"))
res_frozen
```

```python
res_frozen.A = me.A()
res_frozen
```

```python
del res_frozen.Ms
res_frozen
```
